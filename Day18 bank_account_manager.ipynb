```python
# Project Title: Bank Account Manager

# Global variable for transaction tracking
transaction_counter = 0

class BankAccount:
    """A simple class to represent a bank account."""

    def __init__(self, account_holder, balance=0):
        # Instance attributes
        self.account_holder = account_holder
        self.balance = balance
        print(f"üè¶ Account created for {self.account_holder} with initial balance ‚Çπ{self.balance}")

    def deposit(self, amount):
        """Deposit money into the account."""
        global transaction_counter
        if amount > 0:
            self.balance += amount
            transaction_counter += 1
            print(f"‚úÖ Deposited ‚Çπ{amount}. New balance: ‚Çπ{self.balance}")
        else:
            print("‚ö†Ô∏è Deposit amount must be positive.")

    def withdraw(self, amount):
        """Withdraw money if sufficient balance is available."""
        global transaction_counter
        if amount <= 0:
            print("‚ö†Ô∏è Withdrawal amount must be positive.")
        elif amount > self.balance:
            print("‚ùå Insufficient balance! Withdrawal denied.")
        else:
            self.balance -= amount
            transaction_counter += 1
            print(f"üí∏ Withdrawn ‚Çπ{amount}. Remaining balance: ‚Çπ{self.balance}")

    def display_balance(self):
        """Display the current account balance."""
        print(f"üìä {self.account_holder}'s current balance: ‚Çπ{self.balance}")


# Demonstration of Function Arguments, Local/Global Variables, and OOP
def perform_transactions(account, deposit_amt=0, withdraw_amt=0):
    """Function to demonstrate local vs global scope and positional/keyword arguments."""
    print("\n--- Transaction Summary ---")
    local_message = "Processing transactions..."
    print(local_message)  # Local variable
    account.deposit(deposit_amt)
    account.withdraw(withdraw_amt)
    account.display_balance()


# Step 1: Create accounts using positional arguments
account1 = BankAccount("Agnes", 5000)

# Step 2: Call methods using positional and keyword arguments
perform_transactions(account1, deposit_amt=2000, withdraw_amt=1500)

# Step 3: Create another account using keyword arguments
account2 = BankAccount(account_holder="John", balance=3000)

perform_transactions(account2, withdraw_amt=3500)  # Attempt to overdraw
perform_transactions(account2, deposit_amt=1000, withdraw_amt=500)

# Step 4: Display total transactions (global variable)
print("\nüèÅ Total Transactions Made:", transaction_counter)
```
