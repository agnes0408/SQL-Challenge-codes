```python
# Project Title: Retail Inventory Manager

# Step 1: Create a tuple with fixed product categories
product_categories = ("Electronics", "Stationery", "Clothing")

# Step 2: Create a list of products that can be updated dynamically
products = ["Laptop", "Pen", "Shirt", "Charger"]

# Step 3: Ask the user to enter a new product and add it to the list
new_product = input("Enter a new product to add: ")
products.append(new_product)
print(f"{new_product} has been added to the inventory.")

# Step 4: Ask the user to input the index of a product to remove it from the list
print("\nCurrent products with indexes:")
for i, item in enumerate(products):
    print(f"{i}: {item}")

remove_index = int(input("\nEnter the index of the product to remove: "))

# Check for valid index
if 0 <= remove_index < len(products):
    removed_product = products.pop(remove_index)
    print(f"{removed_product} has been removed from the inventory.")
else:
    print("Invalid index. No product removed.")

# Step 5: Display the final list of products and product categories
print("\nüõçÔ∏è Final Inventory Details:")
print(f"Product Categories (Tuple - Immutable): {product_categories}")
print(f"Products (List - Mutable): {products}")
```
