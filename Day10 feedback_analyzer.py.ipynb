```python
# Project Title: Feedback Analyzer

# Step 1: Ask the user to input a sentence (feedback)
feedback = input("Enter your feedback: ")

# Step 2: Count how many words are in the sentence
# The split() method divides the sentence by spaces into a list of words
word_count = len(feedback.split())
print(f"Total number of words: {word_count}")

# Step 3: Check how many times the word “data” appears (case-insensitive)
# Convert the sentence to lowercase before counting
data_count = feedback.lower().count("data")
print(f"The word 'data' appears {data_count} times.")

# Step 4: Extract and print the first 5 and last 5 characters
first_five = feedback[:5]   # slicing from start to 5th character
last_five = feedback[-5:]   # slicing last 5 characters
print(f"First 5 characters: {first_five}")
print(f"Last 5 characters: {last_five}")

# Step 5: Reverse the sentence and print it
reversed_feedback = feedback[::-1]
print(f"Reversed feedback: {reversed_feedback}")
```
