```python
# Project Title: Contact Directory

# Step 1: Create a 2D list where each row holds [Name, Phone, Email]
contacts = [
    ["Alice", "9876543210", "alice@example.com"],
    ["Bob", "9123456780", "bob@example.com"],
    ["Charlie", "9988776655", "charlie@example.com"]
]

# Step 2: Allow the user to add a new contact
print("Add a new contact:")
name = input("Enter name: ").strip().title()
phone = input("Enter phone number: ").strip()
email = input("Enter email address: ").strip()

contacts.append([name, phone, email])
print(f"\n‚úÖ Contact for {name} added successfully!")

# Step 3: Display all contacts sorted alphabetically by name
contacts.sort(key=lambda x: x[0])
print("\nüìá All Contacts (Sorted Alphabetically):")
for contact in contacts:
    print(f"Name: {contact[0]}, Phone: {contact[1]}, Email: {contact[2]}")

# Step 4: Allow the user to search for contacts by (partial) name
search_query = input("\nüîç Enter a name to search (supports partial match): ").strip().lower()
print("\nüîé Search Results:")
found = False
for contact in contacts:
    if search_query in contact[0].lower():  # partial match
        print(f"Name: {contact[0]}, Phone: {contact[1]}, Email: {contact[2]}")
        found = True
if not found:
    print("No matching contacts found.")

# Step 5: Display only the first 3 contacts after sorting
print("\nüìå First 3 Contacts in Directory:")
for contact in contacts[:3]:  # list slicing
    print(f"Name: {contact[0]}, Phone: {contact[1]}, Email: {contact[2]}")
```
