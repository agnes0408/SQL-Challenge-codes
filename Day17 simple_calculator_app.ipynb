```python
# Project Title: Simple Calculator App

# Step 1: Define functions for basic operations

def add(a, b):
    """Return the sum of two numbers."""
    return a + b

def subtract(a, b):
    """Return the difference between two numbers."""
    return a - b

def multiply(a, b):
    """Return the product of two numbers."""
    return a * b

def divide(a, b):
    """Return the division result, handling division by zero."""
    if b == 0:
        return "âŒ Error: Division by zero is not allowed."
    return a / b


# Step 2: Ask the user for input
print("ğŸ§® Welcome to the Simple Calculator App ğŸ§®")
try:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    operation = input("Enter operation (+, -, *, /): ")

    # Step 3: Perform the selected operation
    if operation == '+':
        result = add(num1, num2)
    elif operation == '-':
        result = subtract(num1, num2)
    elif operation == '*':
        result = multiply(num1, num2)
    elif operation == '/':
        result = divide(num1, num2)
    else:
        result = "âŒ Invalid operation selected."

    # Step 4: Use built-in functions
    if isinstance(result, (int, float)):
        result = round(result, 2)  # round result to 2 decimal places
        abs_result = abs(result)   # get absolute value
        power_result = pow(result, 2)  # square the result
        print(f"\nâœ… Result: {result}")
        print(f"ğŸ”¹ Absolute Value: {abs_result}")
        print(f"ğŸ”¹ Result Squared: {power_result}")
    else:
        print(result)

except ValueError:
    print("âš ï¸ Please enter valid numeric values only.")

print("\nğŸ¯ Thank you for using the Simple Calculator App!")
```
