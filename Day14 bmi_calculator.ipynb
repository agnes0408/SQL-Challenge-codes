```python
# Project Title: BMI Calculator

# Step 1: Ask the user to enter weight (kg) and height (m)
try:
    weight = float(input("Enter your weight in kilograms: "))
    height = float(input("Enter your height in meters: "))

    # Step 2: Validate inputs (must be positive)
    if weight <= 0 or height <= 0:
        print("âŒ Please enter positive values for weight and height.")
    else:
        # Step 3: Calculate BMI using formula
        bmi = weight / (height * height)

        # Step 4: Classify BMI using conditional statements
        if bmi < 18.5:
            category = "Underweight"
        elif 18.5 <= bmi < 24.9:
            category = "Normal weight"
        elif 25 <= bmi < 29.9:
            category = "Overweight"
        else:
            category = "Obese"

        # Step 5: Display the result clearly
        print(f"\nðŸ“Š Your BMI is: {bmi:.2f}")
        print(f"ðŸ’¡ Health Category: {category}")

except ValueError:
    print("âš ï¸ Invalid input! Please enter numeric values only.")
```
