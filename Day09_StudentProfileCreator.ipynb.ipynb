### üßë‚Äçüéì Project Title: Student Profile Creator  

**Scenario:**  
You are helping to build a basic student information system for a local school.  
The school wants to store some basic details about each student and let them update their interests.

**Tasks:**  
1. Create variables to store the student‚Äôs name, age, and their favourite subjects they are enrolled in.  
2. Display the student‚Äôs information in a user-friendly format.  
3. Ask the student to enter their favourite subject and add it to their favourite subjects variable.  
4. Display the updated list of subjects.


```python
# Step 1: Create variables
student_name = input("Enter your name: ")
student_age = int(input("Enter your age: "))
favourite_subjects = ["Math", "Science", "English"]
```

```python
# Step 2: Display student information
print("\n--- Student Profile ---")
print(f"Name: {student_name}")
print(f"Age: {student_age}")
print(f"Favourite Subjects: {', '.join(favourite_subjects)}")
```

```python
# Step 3: Ask for a new favourite subject
new_subject = input("\nEnter your favourite subject to add: ")
favourite_subjects.append(new_subject)
```

```python
# Step 4: Display updated list
print("\n--- Updated Student Profile ---")
print(f"Name: {student_name}")
print(f"Age: {student_age}")
print(f"Favourite Subjects: {', '.join(favourite_subjects)}")
```
