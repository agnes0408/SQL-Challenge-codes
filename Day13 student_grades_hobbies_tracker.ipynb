```python
# Project Title: Student Grades & Hobbies Tracker

# Step 1: Create a dictionary with student names as keys and their grades as values
student_grades = {
    "Alice": 85,
    "Bob": 78,
    "Charlie": 92
}

print("ğŸ“ Current Student Grades:")
for name, grade in student_grades.items():
    print(f"{name}: {grade}")

# Step 2: Allow the user to add or update grades for at least 3 students
print("\nâœï¸ Update or Add Grades for Students:")
for i in range(3):
    name = input("Enter student name: ").strip().title()
    grade = int(input(f"Enter grade for {name}: "))
    student_grades[name] = grade  # adds new or updates existing
    print(f"âœ… Grade updated for {name}")

# Display updated dictionary
print("\nğŸ“˜ Updated Student Grades:")
for name, grade in student_grades.items():
    print(f"{name}: {grade}")

# Step 3: Create a set of hobbies and allow user input for at least 3 hobbies
hobbies = set()
print("\nğŸ¨ Enter at least 3 hobbies:")
for i in range(3):
    hobby = input(f"Enter hobby {i+1}: ").strip().capitalize()
    hobbies.add(hobby)

print(f"\nYour hobbies (as a set): {hobbies}")

# Step 4: Convert the set of hobbies into a frozenset and display it
frozen_hobbies = frozenset(hobbies)
print(f"\nâ„ï¸ Frozenset of hobbies (immutable): {frozen_hobbies}")
```
