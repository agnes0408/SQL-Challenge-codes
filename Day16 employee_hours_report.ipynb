```python
# Project Title: Employee Hours Report (Student Exam Scores Analysis)

import random

# Step 1: Create a list of 10 random exam scores (0â€“100) using list comprehension
scores = [random.randint(0, 100) for _ in range(10)]

print("ðŸŽ“ Student Exam Scores Report\n")
print("All Exam Scores:", scores)
print("-" * 40)

# Step 2: Display each score with its index using enumerate
print("ðŸ“‹ Scores with Index:")
for index, score in enumerate(scores):
    print(f"Student {index + 1}: {score}")
print("-" * 40)

# Step 3: Find all scores greater than 75
high_scores = [score for score in scores if score > 75]
print(f"ðŸŒŸ Scores greater than 75: {high_scores if high_scores else 'None'}")
print("-" * 40)

# Step 4: Calculate and display average score
average_score = sum(scores) / len(scores)
print(f"ðŸ“Š Average Score: {average_score:.2f}")
```
