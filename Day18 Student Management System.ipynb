**Studant Management System**



```python
# -------------------------------------------
# Student Management System
# -------------------------------------------

# CONSTANT DATA (Using tuple)
CLASSES = ("Class 1", "Class 2", "Class 3", "Class 4")

# STORAGE FOR STUDENTS
students = []

# ------------------------------------------------
# FUNCTION: Add a student
# ------------------------------------------------
def add_student():
    try:
        name = input("Enter student name: ")
        age = int(input("Enter age: "))
        class_name = input(f"Enter class {CLASSES}: ")

        if class_name not in CLASSES:
            print("Invalid class name!")
            return

        grades = float(input("Enter grade (0–100): "))

        student = {
            "name": name,
            "age": age,
            "class": class_name,
            "grade": grades
        }

        students.append(student)
        print("Student added successfully!")

    except ValueError:
        print("Invalid input! Try again.")


# ------------------------------------------------
# FUNCTION: Delete a student
# ------------------------------------------------
def delete_student():
    name = input("Enter name of student to delete: ")
    for student in students:
        if student["name"].lower() == name.lower():
            students.remove(student)
            print("Student removed!")
            return
    print("Student not found.")


# ------------------------------------------------
# FUNCTION: Update a student
# ------------------------------------------------
def update_student():
    name = input("Enter student name to update: ")

    for student in students:
        if student["name"].lower() == name.lower():
            print("1. Update Age")
            print("2. Update Class")
            print("3. Update Grade")

            choice = int(input("Choose option: "))

            if choice == 1:
                student["age"] = int(input("Enter new age: "))
            elif choice == 2:
                class_name = input(f"Enter class {CLASSES}: ")
                if class_name in CLASSES:
                    student["class"] = class_name
                else:
                    print("Invalid class!")
            elif choice == 3:
                student["grade"] = float(input("Enter new grade: "))
            else:
                print("Invalid option.")

            print("Student updated!")
            return

    print("Student not found.")


# ------------------------------------------------
# FUNCTION: Search student
# ------------------------------------------------
def search_student():
    name = input("Enter student name to search: ")

    for student in students:
        if student["name"].lower() == name.lower():
            print(student)
            return
    print("Student not found.")


# ------------------------------------------------
# FUNCTION: Grade Category
# ------------------------------------------------
def categorize_grade(grade):
    if grade >= 90:
        return "Excellent"
    elif grade >= 75:
        return "Good"
    elif grade >= 50:
        return "Average"
    else:
        return "Poor"


# ------------------------------------------------
# FUNCTION: Show report with sorting
# ------------------------------------------------
def student_report():
    print("1. Sort by Name")
    print("2. Sort by Grade (High to Low)")
    print("3. Sort by Age")

    choice = int(input("Choose sorting option: "))

    if choice == 1:
        sorted_list = sorted(students, key=lambda x: x["name"])
    elif choice == 2:
        sorted_list = sorted(students, key=lambda x: x["grade"], reverse=True)
    elif choice == 3:
        sorted_list = sorted(students, key=lambda x: x["age"])
    else:
        print("Invalid choice.")
        return

    for s in sorted_list:
        print(f"{s['name']} - {s['grade']} - {categorize_grade(s['grade'])}")


# ------------------------------------------------
# FILE HANDLING
# ------------------------------------------------
def save_data():
    import json
    with open("students.json", "w") as f:
        json.dump(students, f)
    print("Data saved!")


def load_data():
    import json
    try:
        with open("students.json", "r") as f:
            data = json.load(f)
            students.extend(data)
        print("Data loaded successfully!")
    except FileNotFoundError:
        print("No previous data found.")


# ------------------------------------------------
# MAIN PROGRAM LOOP
# ------------------------------------------------
load_data()

while True:
    print("\n===== STUDENT MANAGEMENT SYSTEM =====")
    print("1. Add Student")
    print("2. Delete Student")
    print("3. Update Student")
    print("4. Search Student")
    print("5. Student Report")
    print("6. Save and Exit")

    try:
        option = int(input("Choose an option: "))

        if option == 1:
            add_student()
        elif option == 2:
            delete_student()
        elif option == 3:
            update_student()
        elif option == 4:
            search_student()
        elif option == 5:
            student_report()
        elif option == 6:
            save_data()
            print("Exiting program...")
            break
        else:
            print("Invalid option. Try again.")

    except ValueError:
        print("Please enter a valid number.")
```
