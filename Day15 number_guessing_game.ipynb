```python
# Project Title: Number Guessing Game

import random

# Step 1: Randomly choose a number between 1 and 20
secret_number = random.randint(1, 20)
attempts = 0
max_attempts = 5

print("ðŸŽ¯ Welcome to the Number Guessing Game!")
print("I'm thinking of a number between 1 and 20.")
print(f"You have {max_attempts} attempts to guess it.\n")

# Step 2: Loop for up to 5 attempts
while attempts < max_attempts:
    try:
        guess = int(input("Enter your guess: "))

        # Step 3: Increment attempt count only for valid guesses
        attempts += 1

        # Step 4: Compare guess with secret number
        if guess == secret_number:
            print(f"ðŸŽ‰ Congratulations! You guessed the number {secret_number} correctly in {attempts} attempts.")
            break
        elif guess < secret_number:
            print("ðŸ“‰ Too low! Try again.\n")
        else:
            print("ðŸ“ˆ Too high! Try again.\n")

    except ValueError:
        # Handle invalid input (non-integer)
        print("âš ï¸ Invalid input! Please enter a valid number.\n")
        continue  # doesnâ€™t count as an attempt

# Step 5: Else block runs only if the loop completes without a correct guess
else:
    print(f"âŒ Sorry, you've used all {max_attempts} attempts.")
    print(f"The correct number was: {secret_number}")
```
